# Kelvin Probe Force Microscopy (KPFM) Probe Modelling

## 1. Introduction: 

Kelvin probe force microscopy (KPFM) is a scanning probe microscopy measuring technique or a nano-analytics tool used to measure the local contact potential differences between a conducting atomic force microscopy (AFM) tip and the conductive sample which allows to image the local surface potential distribution or work function difference of the sample with sub-nanometer spatial resolution. KPFM has many diverse applications in several fields due to the direct and quantitative measurement of the surface potential distribution of the samples. Since this work function or surface potential distribution strongly affect the chemical and physical properties at the surface of the samples, KPFM has been extensively used as a unique method to characterize the nano-scaled electronic and electrical properties of metallic nanostructures and semiconductor surface and devices, organic material and devices as well as biological materials and devices.



The following figure depicts a block diagram explaining how a general KPFM system operates. An AC voltage with a DC offset bias is applied between a conducting cantilever tip and a conductive ample, which results in generation of an electrostatic force due to the work function difference or potential differece beteween the sample and probe tip. This electrostatic force causes the deflection of the conducting cantilever and this cantilever deflection is monitored by a laser on a position sensitive detector. The frequency component is extracted from the obtained postion sensitive detector signal through a lock-in amplifier. Additionaly, a feedback controller applies the DC voltage to the AFM system until the current AC component is minimized. Consequently, the the contact potential diffence (CPD) is obtained by adjusting the DC component by equalzing to the electstatic force. Thefore, the measurement accuracy of the electrostatic force between the conductive sample and the conductive tip define the scanning solution hence the contact potential difference.

KPFM together with the tapping mode AFM where the probe (cantiliver and tip) is mechanically oscillated with a fixed resonance frequency and fixed setpoint amplitude. There are two major KPFM detection techniques are widley used which are amplitude modulation KPFM (AM-KPFM) and frequency modulation KPFM (FM-KPFM). The AM-KPFM mode KPFM directly detects the electrostatic force by the oscillation of the cantilever, but the FM-KPFM mode KPFM detects the electrostatic force gradient instead of electrostatic force by the frequency shift of the cantilever oscillation. The KPFM measument accuracy and resoluation is largely depend on the the generation and detection techniques of the electrostatic force between an AFM probe and sample. 

## 2. Problem

As we know, in KPFM, an electrostatic force generates while a conductiove AFM probe scans over the conductive sample surface. An AFM probe consist an nanoscaled conductive tip-cone and a large conductive cantilever to support the tip-cone while scanning the sample surface. The tip cone radius kept minimized compare to the cantilever so that the contact area of the tip cone and sample surface is minimized. This eventully maximize the scanning accuracy and resulation of KPFM when the tip sample distance is adjusted to few nanometer. But since the electrostatic force behaves as long range and short range therefore an contribution of the electrostatic force is added to the tip-cone due to the long range effect of the cantilever during the measurments. This additional long range electrostatic forces reduce the scanning resolution and measurment accuracy. Since the meauserment accuracy depends on how the electrostatic force generates by the tip-sample and the cantilever-sample interaction. Probe modeling illuminated the understanding of the scanning resolution and accuracy provided by the two KPFM measurement modes. We utilized a realistic yet straightforward approach to model the probe in an attempt to understand how much of the tip needs to be included to account for maxmize the electrostatic force in AM-KPFM, or maximize electrostatic force gradient in FM-KPFM.

## 3. Mathematical Model Development

Probe modeling starts with the capacitor model for the probe and sample, and approximates the probe as consisting of a micro-cantilever and a tip cone with a point end. The mahematical model have been developed based on the proposed probe. For the model simility, we assume that the KPFM sample has a good conductivity properties and therefore charges are only present on the surfaces. Similarly as a good conductor, the potential is the same all over the probe, and charges are only present on the surfaces. The integrated capacitances of the cantilever and tip cone can be analytically expressed. For this purpose a cantilever and a tip cone model have been developed serately and the electric force and electric force gradient are deduced from the first and second derivatives of the capacitance.


### 3.1 Cantilever Modelling

For cantilever model develpement, first we considered only the cantilever and the sample (shwon in the following figure), ignoring the tip cone effect. For simplicity, we assume the cantilever is parallel to the sample, ignoring the tilt of the cantilever. Recognizing that the electric force at the end of the lever causes more deflection than the same force at the base of the cantilever. 

<div>
<img src="images/model3.PNG",width=500>
</div>

Lets assume the small cantiveler base unit with width $w$ and width $\Delta l $ act as a parallel capacitor to the sample with a distance $H+z$. Therefore, the capacitance of this smaller cantilver is calculated and its normalized in proportion to its length from the base:

\begin{equation}
{\Delta C_{l}}=\epsilon\frac{W\Delta L }{H+z}\left( \frac{l}{L}\right)
\end{equation}

The complete cantilever capacitance is calculated by doing the intregal with the complete length of the cantilever from $0$ to $L$:

\begin{equation}
{C_{Lever}}=\int_0^L\epsilon\frac{W}{H+z}\left( \frac{l}{L}\right)dl = \frac{1}{2}\epsilon\frac{WL}{H+z}
\end{equation}

The electric force from the cantilever is therefore: 

\begin{equation}
{F_{Lever}}=\frac{V^{2}}{4}\epsilon\frac{WL}{\left({H+z}\right)^{2}}               .............(1)
\end{equation}

And the electric force gradient is: 

\begin{equation}
\frac{d{F_{Lever}}}{dz} =\frac{V^{2}}{2}\epsilon\frac{WL}{\left({H+z}\right)^{3}}  .............(2)
\end{equation}


### 3.2 Tipcone Modelling

For the tip cone modelling, a tip cone with height $H$ and tip apex with height $h$ from the sample is considared initially. But for the model simplisity, we assumed that the tip cone is equal to the tip apex hence the $H=h$ in this analysis. The tip cone modelling schamatic is shown in the following figure. 


<div>
<img src="images/model4.PNG",width=500>
</div>

The tip cone capacitance of the small unit of the tip is:

\begin{equation}
{\Delta C_{H}}=\epsilon\frac{2\pi H \tan^{2}\left({\theta}\right)}{H+z}{\Delta H}
\end{equation}

The complete tip cone capacitance is calculated by doing the intregal with the complete area of the tip cone:

\begin{equation}
{C_{H}}=\int_0^H\epsilon\frac{2\pi H \tan^{2} \left({\theta}\right)}{H+z}{d h} = 2 \pi \epsilon z \tan^{2}\left({\theta}\right) \left(\frac{H}{z} - ln \frac{H+z}{z}\right) 
\end{equation}

The electric force from the tip cone is therefore:

\begin{equation}
{F_{tip}}= \pi \epsilon \tan^{2}\left({\theta}\right) \left(\ln \frac{H+z}{z} - \frac{h}{H+z}\right)               .......(3)
\end{equation}

And the electric force gradient is: 

\begin{equation}
\frac{d{F_{tip}}}{dz} =\pi \epsilon \tan^{2}\left({\theta}\right) \frac{1}{z} \left(\frac{H}{H+z}\right)^{2}       .......(4)
\end{equation}

### 5.1 AM-KPFM (Direct Force Measurement) 

####  5.1.1 Cantiliver Modelling and Simulation

##### 5.1.1 .1 Starndard Parameter

In this section, we will simulate the conductive cantiliver-sample electrostatic force interaction based on proposed parameter [1]. The detectable force range by the detetor in AM-KPFM is from few pico newton to several newton. The electrostact force behavior with respect to the tip heigh and cantilever-sample distance will be simulated. The standard parameters considered for this simulation are:

Applied voltage $V = 1V$,
Permittivity of free space $\epsilon_0 = 8.8541878128×10-12 F/m$,
Cantiliver Width $W=30µm$,
Cantiliver Length $L = 225µm$,
tip cone-sample distance  $z = 50 nm$,
Tip-cone angle $Q= 22.5° $

```python
"""Importing the required libraries"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import math
%matplotlib inline
```

Ploting only the cantilever force by varying the tip height and later inspecting exactly for which tip height force become zero.

```python
"""AM-KPFM cantilever force behavior"""

V = 1                # applied voltage 1V
e0 = 8.85*10**-12    # permittivity of the free space
W = 30*10**-6        # width of the cantilever
L = 225*10**-6       # length of the cantilever
z = 50*10**-9        # Tip to sample distance
Q=22.5               # tip angle


def am_cantilever_force(H):
    
    F = ((V**2/4)*e0*W*L)/(H+z)**2              # Cantilever force equation
    
    return F


H1 = np.array(range(1, 1000000))*10**-9         # tip-cone height variation from 1 nm to 100 µm range
H2 = np.array(range(1000, 100000))*10**-9       # tip-cone height variation from 1µm to 100µm range
 
canti_force1 = am_cantilever_force(H1)
canti_force2 = am_cantilever_force(H2)

plt.figure()

fig,ax=plt.subplots(1,2,figsize=(13,5))

ax[0].semilogx(H1*10**6, canti_force1*10**6, label='cantilever',color='red')
ax[0].set_title('Cantilever Force')
ax[0].set_xlabel('Height [µm]')
ax[0].set_ylabel('Forces [µN]')
ax[0].grid(axis='both', which='both')
ax[0].legend()


# PLoting to find out the exact tip height for which force become zero. 

ax[1].semilogx(H2*10**6, canti_force2*10**9, label='cantilever',color='green')
ax[1].set_title('Cantilever Force [Region of Interest]')
ax[1].set_xlabel('Height [µm]')
ax[1].set_ylabel('Forces [nN]')
ax[1].grid(axis='both', which='both')
ax[1].legend()

fig.tight_layout(pad=4.0)

plt.show()
```

From the cantiliver force graph (left figure), the maximum electrostatic force from the cantilever is aproximatily $6µN$ at $1nm$ tip-cone height and the electrostatic force is gradually decreases with increasing the tip-cone height. At the $10µm$ tip-cone height (which is proposed as the standard tip-cone height in [1]) the electrostatic force is shown aproximately zero. Therefore, we simulate the cantilever force from $1µm$ to $100µm$ and called the region of interest (right figure). At $1µm$ tip-cone height the force is aproximately $14nN$ and similarly the electrostatic force is gradually decreases with increasing the tip-cone height.

#### 5.1.2 Tip-cone Modelling and Simulation

##### 5.1.2.1 Starndard Parameter

In this section, we will simulate the tip cone-sample electrostatic force interaction based on proposed parameter [1]. The electrostact force behavior with respect to the tip height and cantiliver-sample distance will be simulated. The standard parameters considered for this simulation are:

Applied voltage $V = 1V$,
Permittivity of free space $\epsilon_0 = 8.8541878128×10-12 F/m$,
Cantiliver Width $W=30µm$,
Cantiliver Length $L = 225µm$,
tip cone-sample distance  $z = 50 nm$,
Tip-cone angle $Q= 22.5° $

```python
"""AM-KPFM tip force behavior"""

V = 1                # applied voltage 1V
e0 = 8.85*10**-12    # permittivity of the free space
W = 30*10**-6        # width of the cantilever
L = 225*10**-6       # length of the cantilever
z = 50*10**-9        # Tip to sample distance
Q=22.5               # tip angle

def am_tip_force(H):
    
    F = np.pi*e0*(np.tan(Q))**2*(np.log((H+z)/z)-(H/(H+z)))    # Tip force at agle 22.5 degree and z=50nm
    
    return F


H1 = np.array(range(1, 100000))*10**-9
H2 = (np.array(range(10, 10000)))*10**-9


tip_force1=am_tip_force(H1)
tip_force2=am_tip_force(H2)


plt.figure()

fig,ax=plt.subplots(1,2,figsize=(13,5))

ax[0].semilogx(H1*10**6, tip_force1*10**12, label='cantilever',color='red')
ax[0].set_title('Tip Force')
ax[0].set_xlabel('Height [µm]')
ax[0].set_ylabel('Forces [pN]')
ax[0].grid(axis='both', which='both')
ax[0].legend()

# PLoting to find out the exact tip height for which force become zero. 

ax[1].semilogx(H2*10**6, tip_force2*10**12, label='cantilever',color='green')
ax[1].set_title('Tip Force [region of interest]')
ax[1].set_xlabel('Height [µm]')
ax[1].set_ylabel('Forces [pN]')
ax[1].grid(axis='both', which='both')
ax[1].legend()

fig.tight_layout(pad=4.0)

plt.show()
```

From the tip force graph (left figure), the electrostatic force is gradually increases with increasing the tip-cone height. At the $10µm$ tip-cone height (which is proposed as the standard tip-cone height in [1]) the electrostatic force is shown aproximately $38pN$. Since our aim is to optimize the tip-cone height sothat the maximum electrostatic forces contribute from the tip cone. Since the tip cone height less than $10nm$ are critical to fabricate. Therefore, we simulate the tip-cone force from $10nm$ to $10µm$ and we will be called the region of interest (right figure).

#### 5.1.3 Total cantilever and tip-cone cotribution

#### 5.1.3.1 Total contribution 

In the earlier sections we have simulated the electrostatic force for the cantilever and tip-cone separately with the sample. Now we will combine the tipcone and the cantilever and will simulate the total force behavior when cantilever and tip-cone are as a single structure.

Let's see the cantilever force, tip-cone force and total force behavior in a single plot

```python
V = 1                # applied voltage 1V
e0 = 8.85*10**-12    # permittivity of the free space
W = 30*10**-6        # width of the cantilever
L = 225*10**-6       # length of the cantilever
z = 50*10**-9        # Tip to sample distance
Q=22.5               # tip angle


def am_total_force(H):
    
    cantilever_force = ((V**2/4)*e0*W*L)/(H+z)**2
    tip_force = np.pi*e0*(np.tan(Q))**2*(np.log((H+z)/z)-(H/(H+z)))
    total_force = cantilever_force + tip_force
                       
    return (cantilever_force,tip_force,total_force)
                       
H = (np.array(range(10, 100000)))*10**-9
am_force = am_total_force(H)
                       
plt.semilogx(H*10**9, am_force[0]*10**6, label='cantilever force',color='red')
plt.semilogx(H*10**9, am_force[1]*10**6, label='tip force',color='green')
plt.semilogx(H*10**9, am_force[2]*10**6, label='total force',color='blue')
plt.title('Total Force')
plt.legend()
plt.xlabel('Height [nm]')
plt.ylabel('Forces [µN]')
plt.grid(axis='both', which='both')
plt.show()
```

From the total force graph it is clearly seen that the total force overlap with the cantilver force which means that the total force is dominated by the cantilier force in the AM-KPFM.

```python
# Tip and cantilever force percentage plot

V = 1                # applied voltage 1V
e0 = 8.85*10**-12    # permittivity of the free space
W = 30*10**-6        # width of the cantilever
L = 225*10**-6       # length of the cantilever
z = 50*10**-9        # Tip to sample distance
Q=22.5               # tip angle


def am_percent_contribution(H):
    
    catilever_force = (((V**2/4)*e0*W*L)/(H+z)**2)
    tip_force = np.pi*e0*(np.tan(Q))**2*(np.log((H+z)/z)-(H/(H+z)))
    total_force = catilever_force + tip_force
    
    cantilever_contribution = 100*(catilever_force/total_force)
    tip_contribution = 100*(tip_force/total_force)
    
    return(cantilever_contribution,tip_contribution)

H = (np.array(range(1000, 100000)))*10**-9
H1 = (np.array(range(10, 100)))*10**-6

percent = am_percent_contribution(H)
total_force = am_total_force(H1)

"""plt.semilogx(H*10**6, percent[0], label='cantilever',color='red')
plt.semilogx(H*10**6, percent[1], label='tip',color='green')

plt.semilogx(H*10**9, am_force[0]*10**6, label='cantilever force',color='red')
plt.semilogx(H*10**9, am_force[1]*10**6, label='tip force',color='green')

plt.title('Tip-cantilever percentage Forces')
plt.legend()
plt.xlabel('Height [µm]')
plt.ylabel('Forces contribution [%]')
plt.grid(axis='both', which='both')
plt.show()"""

plt.figure()

fig,ax=plt.subplots(1,2,figsize=(13,5))

ax[0].semilogx(H*10**6, percent[0], label='cantilever',color='red')
ax[0].semilogx(H*10**6, percent[1], label='tip',color='green')
ax[0].set_title('Tip Force')
ax[0].set_xlabel('Height [µm]')
ax[0].set_ylabel('Force Contribution[%]')
ax[0].grid(axis='both', which='both')
ax[0].legend()

# PLoting to find out the exact tip height for which force become zero. 

ax[1].semilogx(H1*10**6, total_force[0]*10**12, label='cantilever force',color='red')
ax[1].semilogx(H1*10**6, total_force[1]*10**12, label='tip force',color='green')
ax[1].set_title('Tip Force [region of interest]')
ax[1].set_xlabel('Height [µm]')
ax[1].set_ylabel('Forces [pN]')
ax[1].grid(axis='both', which='both')
ax[1].legend()

fig.tight_layout(pad=4.0)

plt.show()
```

From the tip-cantilever percentage, we can see that at standard $10µm$ tip height, the tip force contributes only $20$% (corresponding quantitive force is $40pN)$ while scanning the sample whereas $80$% (corresponding   quantitive force is $150pN) $ forces are conributes from the cantilever. ####  5.1.3.3 Tip- sample distance variation on percentage contribution

In this sub-section, we will simulate and analyzed the tip-sample distance variation on the probe modelling to optimize the tip force contribution. Since it is critical to achieve the tip sample distnce less than $z=5nm$. Therefore, we considered few standard tip sample distance such as $z=5nm$, $z=50nm$, $z=100nm$, $z=150nm$.    

```python
V = 1                # applied voltage 1V
e0 = 8.85*10**-12    # permittivity of the free space
W = 30*10**-6        # width of the cantilever
L = 225*10**-6       # length of the cantilever
#z = 50*10**-9        # Tip to sample distance
Q=22.5               # tip angle

def percent_contribution_z(H,z):
    
    catilever_force = (((V**2/4)*e0*W*L)/(H+z)**2)
    tip_force = np.pi*e0*(np.tan(Q))**2*(np.log((H+z)/z)-(H/(H+z)))
    total_force = catilever_force + tip_force
    
    canti_contribution = 100*(catilever_force/total_force)
    tip_contribution = 100*(tip_force/total_force)
    
    return(canti_contribution,tip_contribution)

H = (np.array(range(1000, 100000)))*10**-9

z1=[5,50,100,150]

for height in z1:
    
    percent = percent_contribution_z(H,height*10**-9)
    
    plt.semilogx(H*10**6, percent[0],label='Tip z= %s nm' % height)
    plt.semilogx(H*10**6, percent[1],label='canti z= %s nm' % height)

plt.title('Tip-cantilever percentage Forces')
plt.legend()
plt.xlabel('Height [µm]')
plt.ylabel('Forces contribution [%]')
plt.grid(axis='both', which='both')
plt.show()
```

We can see from the simulation results, at standard tip height $10µm$ with standard tip-sample distance $z=50nm$ still the cantilever contribution dominates (80% from the cantiliver and 20% from the tip force. At smaller tip-sample distance $z=5nm$, still require $16µm$ of tip height to have an equal tip and cantilever force contribution whereas with larger $z=150nm$ need the larger tip height to have have an equal tip and cantilever force contribution. Although the tip cone height reduced few micro-meter for smaller tip sample distance  but  there is no significance effect to do the measurement with the different tip-sample distance.

### 5.2 FM-KPFM  (Force Gradient Measurement) 

####  5.2.1 Cantiliver Modelling and Simulation

##### 5.2.1.1 Starndard Parameter

In this section, we will simulate the tip-sample electrostatic force interaction based on proposed parameter [1] and using the equation (2). The electrostact force behavior with respect to the tip height will be simulated. The standard parameters considered for this simulation are:

Applied voltage $V = 1V$,
Permittivity of free space $\epsilon_0 = 8.8541878128×10-12 F/m$,
Cantiliver Width $W=30µm$,
Cantiliver Length $L = 225µm$,

```python
"""FM-KPFM cantilever force behavior"""

V = 1                # applied voltage 1V
e0 = 8.85*10**-12    # permittivity of the free space
W = 30*10**-6        # width of the cantilever
L = 225*10**-6       # length of the cantilever
z = 50*10**-9        # Tip to sample distance
Q=22.5               # tip angle

def fm_canti_force(H):
    
    dF_dt = (((V**2/2)*(e0*(W*L)))/((H+z)**3))   # Cantilever force equation in micro-newton range
    
    return dF_dt


H1 = np.array(range(1, 100000))*10**-9     # tip-cone height variation from 1 nano-meter to 100 micro-meter range
H2 = np.array(range(1000, 100000))*10**-9         # tip-cone height variation from 1µm to 100µm range
 
canti_force1 = fm_canti_force(H1)
canti_force2 = fm_canti_force(H2)

plt.figure()

fig,ax=plt.subplots(1,2,figsize=(13,5))

ax[0].semilogx(H1*10**6, canti_force1, label='cantilever',color='red')
ax[0].set_title('Cantilever Force')
ax[0].set_xlabel('Height [µm]')
ax[0].set_ylabel('Forces Gradient [N]')
ax[0].grid(axis='both', which='both')
ax[0].legend()


ax[1].semilogx(H2*10**6, canti_force2*10**3, label='cantilever',color='green')
ax[1].set_title('Cantilever Force (Region of Interest) ')
ax[1].set_xlabel('Height [µm]')
ax[1].set_ylabel('Forces Gradient [mN]')
ax[1].grid(axis='both', which='both')
ax[1].legend()

fig.tight_layout(pad=3.0)

plt.show()
```

The cantilver electrostatc force behaviour in the FM-KPFM are similiar to the cantiliver force behavior in AM-KPFM.From cantiliver force graph (left figure), the maximum electrostatic force from the cantilever is aproximatily $200N$ at $1nm$ tip-cone height and the electrostatic force is gradually decreases with increasing the tip-cone height. #### 5.2.2 Tip-cone Modelling and Simulation

##### 5.1.2.1 Starndard Parameter

In this sub-section, we will simulate the tip cone-sample electrostatic force interaction based on proposed parameter [1] and the equation (4). The electrostact force behavior with respect to the tip height will be simulated. The standard parameters considered for this simulation are:

Applied voltage $V = 1V$,
Permittivity of free space $\epsilon_0 = 8.8541878128×10-12 F/m$,
Cantiliver Width $W=30µm$,
Cantiliver Length $L = 225µm$,
tip cone-sample distance  $z = 50 nm$,
Tip-cone angle $Q= 22.5° $

```python
V = 1                # applied voltage 1V
e0 = 8.85*10**-12    # permittivity of the free space
W = 30*10**-6        # width of the cantilever
L = 225*10**-6       # length of the cantilever
z = 50*10**-9        # Tip to sample distance
Q=22.5               # tip angle

def fm_tip_force(H):
    
    F = (np.pi*e0*(np.tan(Q)**2)**2*((1/z)*(H/(H+z)**2)))    # Tip force at agle 22.5 degree and z=50nm
    
    return F


H = (np.array(range(1, 100000))*10**-9)
H2 = (np.array(range(10, 10000)))*10**-9


tip_force1=fm_tip_force(H1)
tip_force2=fm_tip_force(H2)


plt.figure()

fig,ax=plt.subplots(1,2,figsize=(13,5))

ax[0].semilogx(H1*10**6, tip_force1, label='Tip-cone',color='red')
ax[0].set_title('Tip Force')
ax[0].set_xlabel('Height [µm]')
ax[0].set_ylabel('Forces Gradient [N]')
ax[0].grid(axis='both', which='both')
ax[0].legend()


ax[1].semilogx(H2*10**6, tip_force2, label='Tip-cone',color='green')
ax[1].set_title('Tip Force')
ax[1].set_xlabel('Height [µm]')
ax[1].set_ylabel('Forces Gradient [N]')
ax[1].grid(axis='both', which='both')
ax[1].legend()

fig.tight_layout(pad=3.0)

plt.show()
```

From the simulation results, we can see the tip cone-sample electrostatic force behaviour in FM-KPFM is completely different than the tip cone force in the AM-KPFM.From the tip force graph (left figure), the electrostatic force is basically following the tip cone shape. The reason is that we simplified the electrostatic force model as capacitaor model where the capacitor is form as ring capacitor with the sample. Moreover, the force is large enough to represents the high resolution scanning images.  
At the $10µm$ tip-cone height, the electrostatic force is shown aproximately $10N$. #### 5.2.3 Total cantilever and tip-cone cotribution

#### 5.2.3.1 Total contribution 

In the earlier sections we have simulated the electrostatic force for the cantilever and tip-cone separately with the sample. Now we will combine the tipcone and the cantilever and will simulate the total force behavior when cantilever and tip-cone are as a single structure.

```python
V = 1                # applied voltage 1V
e0 = 8.85*10**-12    # permittivity of the free space
W = 30*10**-6        # width of the cantilever
L = 225*10**-6       # length of the cantilever
z = 50*10**-9        # Tip to sample distance
Q=22.5               # tip angle

def fm_total_force(H):
    
    catilever_force = (((V**2/2)*(e0*(W*L)))/((H+z)**3))
    tip_force = (np.pi*e0*(np.tan(Q)**2)**2*((1/z)*(H/(H+z)**2)))
    total_force = catilever_force + tip_force
                       
    return (catilever_force,tip_force,total_force)
                       
H = (np.array(range(1, 100000)))*10**-9
fm_force = fm_total_force(H)
                       
plt.semilogx(H*10**9, fm_force[0], label='cantilever force',color='red')
plt.semilogx(H*10**9, fm_force[1], label='tip force',color='green')
plt.semilogx(H*10**9, fm_force[2], label='total force',color='blue')
plt.title('Total Forces Gradient')
plt.legend()
plt.xlabel('Height [nm]')
plt.ylabel('Forces Gradient [µN]')
plt.grid(axis='both', which='both')
plt.show()

```

From the FM-KPFM total force simulation, it is clearly seen that the total force does not follow anymore the cantilever force like in the  AM-KPFM but the force is follows the the tip cone force which means that the total force is dominated by the tip cone force in the AM-KPFM.
